```python
import numpy as np
from matplotlib.pyplot import *
%matplotlib inline
import math

def distance(x1, x2, y1, y2, z1, z2):
    # This function describes the absolute distance between points (x1,y1,z1) and (x2, y2, z2)
    return np.sqrt(((x1 - x2) ** 2 + (y1 - y2) ** 2 + (z1 - z2) ** 2))

MWData = np.genfromtxt("Orbit_MW.txt", dtype = None, names = True, skip_header = 0) # Opens Milky Way file and reads its data
M31Data = np.genfromtxt("Orbit_M31.txt", dtype = None, names = True, skip_header = 0) # Opens M31 file and reads its data
M33Data = np.genfromtxt("Orbit_M33.txt", dtype = None, names = True, skip_header = 0) # Opens M33 file and reads its data

plot(MWData['t'], distance(MWData['x'], M31Data['x'], MWData['y'], M31Data['y'], MWData['z'], M31Data['z']), 'r')
xlabel('Time (Ga)')
ylabel('Distance (kpc)')
title('Separation of the Milky Way and M31 over time')
```

```python
plot(M31Data['t'], distance(M33Data['x'], M31Data['x'], M33Data['y'], M31Data['y'], M33Data['z'], M31Data['z']), 'r')
xlabel('Time (Ga)')
ylabel('Distance (kpc)')
title('Separation of M33 and M31 over time')
```

```python
plot(MWData['t'], distance(MWData['vx'], M31Data['vx'], MWData['vy'], M31Data['vy'], MWData['vz'], M31Data['vz']), 'r')
xlabel('Time (Ga)')
ylabel('Relative velocity(km/s)')
title('COM Velocity difference of the Milky Way and M31 over time')
```

```python
plot(M33Data['t'], distance(M33Data['vx'], M31Data['vx'], M33Data['vy'], M31Data['vy'], M33Data['vz'], M31Data['vz']), 'r')
xlabel('Time (Ga)')
ylabel('Relative velocity(km/s)')
title('COM Velocity difference of M33 and M31 over time')
```

1) M31 and the Milky Way will experience 3 close encounters in the future before merging on the third close encounter, visible in their relative position graph.

2) Relative position minima are generally well correlated with maxima in relative velocity, until the final merge, when the relative velocities of M31 and the Milky Way vanish.

3) M31 and the Milky Way merge at roughly 6.2 Ga. At this time, the frequency of M31's velocity jumps increases drastically, as M31's velocity amplitude (relative to the Milky Way) drops precipitously, down to a minimal shift in velocity.

4) After 6 Ga, M33 appears to experience approximately a 20% drop in orbital distance every 750 Ma. If we count a "merge" as the moment when the amplitude of the distance drops below, say, 20 kpc, then M33 will merge with the merged galaxy in about 4.4 Ga from the start time (so 10.4 Ga overall). We don't observe this because our data has M33 a bit further away than the estimated distance in the homework.
